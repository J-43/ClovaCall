```python
import librosa
import IPython.display as ipd
```

```python
y, sr = librosa.load("./zeroth_korean/train_data_01/003/106/106_003_0098.wav", mono=True, sr=16000)

print(librosa.get_duration(y, sr=sr))
```

```python
ipd.Audio(y, rate=sr) # load a local WAV file
```

```python
yt, index = librosa.effects.trim(y, top_db=25)

print(librosa.get_duration(yt, sr=sr))
```
